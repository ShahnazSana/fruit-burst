**Bar Chart**

```
membership_counts = gym['abonoment_type'].value_counts()
plt.bar(membership_counts.index, membership_counts.values)
plt.title('Gym Membership Types')
plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7)
plt.savefig("/content/drive/MyDrive/Math_score.png",dpi=300, bbox_inches='tight')
plt.savefig("/content/drive/MyDrive/Math_score.jpg",dpi=300, bbox_inches='tight')
plt.savefig("/content/drive/MyDrive/Math_score.pdf",dpi=300, bbox_inches='tight')
plt.show()
```

**Histogram**

```
plt.hist(gym['avg_time_in_gym'], bins=10, edgecolor='orange')
plt.title('Histogram of Average Time Spent in Gym')
plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7)
plt.savefig("/content/drive/MyDrive/Math_score.png",dpi=300, bbox_inches='tight')
plt.savefig("/content/drive/MyDrive/Math_score.jpg",dpi=300, bbox_inches='tight')
plt.savefig("/content/drive/MyDrive/Math_score.pdf",dpi=300, bbox_inches='tight')
plt.show()
```

**Scatter**

```
plt.scatter(gym['Age'], gym['visit_per_week'])
plt.title('Age vs Visits per Week')
plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7)
plt.savefig("/content/drive/MyDrive/Math_score.png",dpi=300, bbox_inches='tight')
plt.savefig("/content/drive/MyDrive/Math_score.jpg",dpi=300, bbox_inches='tight')
plt.savefig("/content/drive/MyDrive/Math_score.pdf",dpi=300, bbox_inches='tight')
plt.show()
```

**Box plot**

```
sns.boxplot(data=gym, x='gender', y='Age')
plt.title('Box Plot of Age by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7)
plt.savefig("/content/drive/MyDrive/Math_score.png",dpi=300, bbox_inches='tight')
plt.savefig("/content/drive/MyDrive/Math_score.jpg",dpi=300, bbox_inches='tight')
plt.savefig("/content/drive/MyDrive/Math_score.pdf",dpi=300, bbox_inches='tight')
plt.show()
```

**Count Polt**

```
sns.countplot(data=gym, x='days_per_week')
plt.title('Count Plot of Gym Visits per Week')
plt.xlabel('Days per Week')
plt.ylabel('Count')
plt.grid(True, which="both", linestyle="-", linewidth=0.8, alpha=0.7)
plt.savefig("/content/drive/MyDrive/Math_score.png",dpi=300, bbox_inches='tight')
plt.savefig("/content/drive/MyDrive/Math_score.jpg",dpi=300, bbox_inches='tight')
plt.savefig("/content/drive/MyDrive/Math_score.pdf",dpi=300, bbox_inches='tight')
plt.show()
```

**Bar Plot**

```
sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
plt.title('Average Visits per Week by Membership Type')
plt.xlabel('Membership Type')
plt.ylabel('Average Visits per Week')
plt.grid(True, which="both", linestyle="-", linewidth=0.8, alpha=0.7)
plt.savefig("/content/drive/MyDrive/Math_score.png",dpi=300, bbox_inches='tight')
plt.savefig("/content/drive/MyDrive/Math_score.jpg",dpi=300, bbox_inches='tight')
plt.savefig("/content/drive/MyDrive/Math_score.pdf",dpi=300, bbox_inches='tight')
plt.show()
```
